```python
import pandas as pd

def existing(file_name): # функция проверки существования таблицы
    if file_name in DBs.keys():
        return True
    return False

DBs = dict() # словарь таблиц
filename = input() # имя входного файла

# Далее идёт проверка формата файла
if filename.split('.')[-1] == 'csv':
    if existing(filename[:-4]) == False:
        DBs[filename[:-4]] = pd.read_csv(filename) # создание новой таблицы в словре
    else:
        DBs[filename[:-4]].join(pd.read_csv(filename), how='outer') # объединение через outer_join

elif filename.split('.')[-1] == 'xlsx':
    if existing(filename[:-5]) == False:
        DBs[filename[:-5]] = pd.read_excel(filename)
    else:
        DBs[filename[:-5]].join(pd.read_excel(filename), how='outer')
```
